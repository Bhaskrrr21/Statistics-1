5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF) ?
- The PDF provides the density of probabilities for specific values, while the CDF provides cumulative probabilities up to a certain value.
- The PDF is used to describe the distribution of a continuous random variable and to calculate probabilities over intervals
- The CDF is used to determine the probability that a random variable takes on a value less than or equal to a specific point
- The PDF is typically represented as a curve on a graph, where the x-axis represents the values of the random variable and the y-axis represents the density of probability.
- The CDF is represented as a non-decreasing curve that starts at 0 and approaches 1. It shows the cumulative probability up to a certain value

3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution ?

```python
def bernoulli_pdf(x, p):
    """
    Calculate the Bernoulli Probability Density Function (PDF).
    Parameters:
    - x (int): Outcome (0 or 1)
    - p (float): Probability of success (0 ‚â§ p ‚â§ 1)

    Returns:
    - float: Probability P(X = x)
    """
    if x == 1:
        return p
    elif x == 0:
        return 1 - p
    else:
        raise ValueError("x must be 0 or 1 in a Bernoulli distribution")

# Example usage
p = 0.7  # Probability of success
print(bernoulli_pdf(1, p))  # Output: 0.7 (P(X=1))
print(bernoulli_pdf(0, p))  # Output: 0.3 (P(X=0))
```

14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution ?

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Function to calculate Z-score
def calculate_z_score(data, mean=None, std_dev=None):
    if mean is None:
        mean = np.mean(data)
    if std_dev is None:
        std_dev = np.std(data, ddof=1)  # Sample standard deviation
    return [(x - mean) / std_dev for x in data]

def plot_z_scores(data):
    z_scores = calculate_z_score(data)

    x = np.linspace(-4, 4, 100)
    y = stats.norm.pdf(x, 0, 1)

    plt.plot(x, y, 'b', linewidth=2, label='Standard Normal Distribution')
    for z in z_scores:
        plt.axvline(z, color='r', linestyle='dashed', alpha=0.5)

    plt.xlabel("Z-score")
    plt.ylabel("Density")
    plt.title("Z-scores Compared to Standard Normal Distribution")
    plt.legend()
    plt.grid()
    plt.show()

    return z_scores

# Example
data = [45, 50, 55, 60, 65]
z_scores = plot_z_scores(data)
print("Z-scores:", z_scores)
```

15. Implement hypothesis testing using Z-statistics for a sample dataset ?

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Function to calculate Z-score
def calculate_z_score(x, mean, std_dev):
    return (x - mean) / std_dev

# Function to perform hypothesis testing using Z-statistics
def hypothesis_test(sample_mean, population_mean, std_dev, sample_size, alpha=0.05):
    z_stat = (sample_mean - population_mean) / (std_dev / np.sqrt(sample_size))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

    print(f"Z-statistic: {z_stat:.2f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis (H0): Significant difference detected.")
    else:
        print("Fail to reject the null hypothesis (H0): No significant difference detected.")

    return z_stat, p_value

# Example
sample_mean = 55
population_mean = 50
std_dev = 10
sample_size = 30
hypothesis_test(sample_mean, population_mean, std_dev, sample_size)
```

16. Create a confidence interval for a dataset using Python and interpret the result ?

```python
import numpy as np
import scipy.stats as stats

def calculate_confidence_interval(data, confidence_level=0.95):

    data = np.array(data)

    # Calculate the sample mean and standard error
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    n = len(data)  # Sample size
    standard_error = sample_std / np.sqrt(n)

    critical_value = stats.t.ppf((1 + confidence_level) / 2, df=n-1)  # t-distribution

    # Calculate the margin of error
    margin_of_error = critical_value * standard_error

    # Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example
if __name__ == "__main__":

    data = [10,20,30,40,50]

    # Calculate the confidence interval
    confidence_level = 0.95
    ci_lower, ci_upper = calculate_confidence_interval(data, confidence_level)

    print(f"Confidence Interval ({confidence_level*100:.0f}%): [{ci_lower:.2f}, {ci_upper:.2f}]")
```

17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean ?

```python
import numpy as np
import scipy.stats as stats

# Function to generate data from a normal distribution
def generate_normal_data(mean, std_dev, size):
    return np.random.normal(mean, std_dev, size)

# Function to calculate confidence interval
def confidence_interval(sample, confidence=0.95):
    mean = np.mean(sample)
    std_err = stats.sem(sample)  # Standard error of the mean
    margin = stats.t.ppf((1 + confidence) / 2, df=len(sample)-1) * std_err
    ci = (mean - margin, mean + margin)
    print(f"{confidence*100:.1f}% Confidence Interval: {ci}")
    return ci

# Generate data
np.random.seed(42)
data = generate_normal_data(50, 10, 30)

# Calculate and interpret confidence interval
ci = confidence_interval(data)
print(f"The true mean is likely within {ci[0]:.2f} to {ci[1]:.2f} with 95% confidence.")
```

23. Implement Z-score normalization on a dataset and visualize its transformation ?

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def generate_data(mean, std_dev, size):

    return np.random.normal(loc=mean, scale=std_dev, size=size)

def z_score_normalization(data):

    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)
    normalized_data = (data - mean) / std_dev
    return normalized_data

def plot_data(original_data, normalized_data):

    plt.figure(figsize=(12, 6))

    # Plot original data
    plt.subplot(1, 2, 1)
    sns.histplot(original_data, bins=30, kde=True, color='blue', alpha=0.6)
    plt.title('Original Data Distribution')
    plt.xlabel('Value')
    plt.ylabel('Frequency')

    # Plot normalized data
    plt.subplot(1, 2, 2)
    sns.histplot(normalized_data, bins=30, kde=True, color='orange', alpha=0.6)
    plt.title('Z-Score Normalized Data Distribution')
    plt.xlabel('Z-Score')
    plt.ylabel('Frequency')

    plt.tight_layout()
    plt.show()

# Example
if __name__ == "__main__":

    mean = 100
    std_dev = 15
    size = 1000
    original_data = generate_data(mean, std_dev, size)

    # Normalize the dataset
    normalized_data = z_score_normalization(original_data)

    plot_data(original_data, normalized_data)
```
